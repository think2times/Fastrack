```python
#------R4-2----------------------
def power(x,n):
    if n==0:
        return 1
    else:
        return x*power(x, n-1)

"""
For this implementation (x=2, n=5) , the trace is

power(2, 5)    returns 16*2  = 32
    power(2,4)       returns 8*2 = 16
        power(2,3)     returns 4*2 = 8
            power (2,2)     returns 2*2 = 4
                power (2,1)   returns 2*1 = 2
                    power (2,0) returns 1


The arrows would trace downwards for the calls and upwards for the returns as in the textbook

"""


print(power(2,5))
```

```python
#-----------R4-4---------------------
def reverse(S, start, stop):
    if start < stop-1:
        S[start], S[stop-1] = S[stop-1], S[start]
        reverse(S, start+1, stop-1)
        
"""
calling reverse on S = [4, 3, 6, 2, 6] (reverse(S, 0, 4) provides:

reverse(S, 0, 5)   return None (S is [6, 2, 6, 3, 4] at the return)
    produces [6, 3, 6, 2, 4]
    recurse(S, 1, 4)    return None (S is [6, 2, 6, 3, 4] at the return)
        produces [6, 2, 6, 3, 4]
        recurses(S, 2, 3)     return None  (S is [6, 2, 6, 3, 4])
            

After everything, S is now (S is [6, 2, 6, 3, 4]).  Note that this function changes the original array

Note that in this recursion, the swaps happen before the next recursion is called, and the return is none


"""

S = [4, 3, 6, 2, 6]
reverse(S, 0, 5)
print(S)
```

```python
#----------R4-7---------------------
"""
Note, we can solve this by knowing that a number can be decomposed into:
a*1e0 + a2*1e1 + a3*1e2, etc...

So 123 is
1*1 + 2*10 + 3*100

"""


def str_2_int(string, index = 0):
    length = len(string)
    if index == length-1:
        return int(string[index])
    else:
        return int(string[index])*10**(length-index-1) + str_2_int(string, index+1)


"""
This runs in 0(n) time and space

""" 
    
ans = str_2_int('12342543')
print (ans, type(ans))    
```

```python
#---------------R4-8---------------------
import math

def isabel_method(A):
    assert math.log(len(A),2)%1 == 0, 'Your array must be a length that is a power of 2'
    if len(A) == 1:
        return A[0]
    else:
        B = [None]*(len(A)//2)
        for i in range(len(B)):
            B[i] = A[2*i] + A[2*i-1]           
        return isabel_method(B)

"""
The recursion will be called log(n) times, since the array is divided in half each time


Each call will require you to access all of the remaining elements, which means it will be
n + n/2 + n/4 + ...

This is equal to n* Sum(1/2**i), which is <2n

Therefore, the running time is O(n), although a much slower implementation than just adding them up element by element

"""
    
                
print(isabel_method([0, 1,2,3,4,5,6,7]))
try:
    print(isabel_method([1,2,3,4,5,6,7]))
except Exception as e:
    print(e)
```

```python
#------------C4-11----------------------
"""
Note, the easiest way is to use a set or dictionary, but I'm not sure we're allowed to do that for this assignment 
(Otherwise it's easy to do this in O(n))

Based on the hint, we are recommended to check if the first element is different from the rest, which 
would produce n-1 + n-2 + .... 1 comparisons, which would add up to  (n)(n+1)/2- n = (n-1)(n)/2
which is O(n)

"""

def unique_no_dictionary(S, index = 0):
    if index == len(S)-1:
        return True    #The trivial case is true
    else:
        unique = True
        for i in range(index+1, len(S)):
            if S[index] == S[i]: unique = False
        #Note, this return will short-circuit if unique is False, therefore saving time
        return unique and unique_no_dictionary(S, index+1)
    
for S in [[1,2,3,4,5,6], 
          [3,4,5,2,3,4,6,7],
          [234,654,32,543,652]
         ]:
    print(unique_no_dictionary(S))
```

```python
#---------C4-14-----------------------
import time
from IPython.display import clear_output

class TowerOfHanoi():
    def __init__(self, n=4):
        self._n = n
        self._array = [[], [], list(reversed(range(n)))]
        self._lengths = [0, 0, n]
        
    def draw_towers(self):
        rows = []
        rows.append(['\t  1\t', '\t  2\t', '\t  3\t'])
        rows.append(['\t_____\t', '\t_____\t', '\t_____\t'])
        for i in range(max(self._lengths)):
            row = []
            for j in range(3):
                if i<self._lengths[j]:
                    row.append('\t  ' + str(self._array[j][i]) + '\t')
                else:
                    row.append('\t  \t')
            rows.append(row)
        
        
        for r in reversed(rows):
            print(''.join(r))
            
    def __getitem__(self, index):
        return self._array[index]
    
    def pop(self, index):
        self._lengths[index] -= 1
        return self[index].pop()
    
    
    def getlen(self):
        return self._lengths
    
    
    def __setitem__(self, index, value):
        if self[index] and self[index][-1] < value:
            raise ValueError (f'Illegal move.  Cannot place block with size {value} on block {self[index][-1]}')
        else: 
            self[index].append(value)
            self._lengths[index] += 1
    
    def _move_stack(self, n_disks, start_peg, help_peg, target_peg):
        time.sleep(0.5)
        clear_output()
        self.draw_towers()
        
        if n_disks == 1:
            self._count += 1
            value = self.pop(start_peg)
            try:
                self[target_peg] = value
            except Exception as e:
                print(e)
                self[start_peg]
        
        else:
            #Move the upper stack to the helper peg
            self._move_stack(n_disks-1, start_peg, target_peg, help_peg)
            #Move the lowest item to the target peg
            self._move_stack(1, start_peg, help_peg, target_peg)
            #Move the upper stack to the target peg
            self._move_stack(n_disks-1, help_peg, start_peg, target_peg)
            
    
    def solve_hanoi(self):
        self._count = 0
        self._move_stack(self.getlen()[2], 2, 1, 0)
        
        time.sleep(0.5)
        clear_output()
        self.draw_towers()
        
        print(f'\nThis took a total of {self._count} moves!')
        
        
            
t = TowerOfHanoi(5)
t.solve_hanoi()
```

```python
#---------------C4-15-----------------------------
"""
Here, we want combinations, not permutations


We rely on the addition of an empty set value 'UNK'

For each stage of the recursion, we add one more value from the sequence
and for each current item either add another UNK or that value

The final layer of that tree will contain exactly one copy of each unique set

"""
UNK = chr(1000)



def sub_rec(U, S):
    """
    U is the current set
    S is the remaining sequence
    """
    if len(S) == 0:
        print('{', str([x for x in U if x!= UNK])[1:-1], '}')
        
    else:
        val = S.pop()
        U.append(UNK)
        sub_rec(U, S)
        U.pop()
        
        U.append(val)
        sub_rec(U, S)
        U.pop()
        S.append(val)

def print_subsets (U):
    sub_rec([], list(U))
    
    
print_subsets({1,2,3,4,5})
    
    
    
```

```python
#--------------C4-17------------------
"""
Note: if a sequence is a palindrome and has an odd number of characters, the middle character has no duplicate
Otherwise, the middle two characters should be identical

"""

def palindrome_checker(S, index = 0):
    l = len(S)
    if l == 1:
        return True
    
    elif index >= l//2:
        #You actually want the (last index + 1)//2, but since the last index is l-1, it becomes (l+1-1)//2, or just l//2
        return S[index] == S[l//2]   #For odd this will be the same index, for even it will not be
      
    else:
        #Note, you want to put the terms in this order to shortcircuit the recursive check.  If it's going to fail, it's
        #likely to fail around the first few indices in most normal cases
        return (S[index] == S[l-index-1]) and palindrome_checker(S, index+1)
    
    
    
strings_list = ['racecar', 'gohangasalamiimalasagnahog', 'notapalindrome', 'bob', 'a', 'hat']

for s in strings_list:
    print (palindrome_checker(s), s)
```

```python
#---------------C4-18------------
"""
Two approaches:
1) Actually count all the vowell and consonants and return two values
2) Have a common counter that increments if it sees a consonant and decrements if it sees a vowell
    It's sign will tell you which is more.
    I'll fo the second because it's more unique

"""

VOWELLS = {'a','e','i','o','u'}

def c_or_v(S, index = 0):
    a = -1 if S[index] in VOWELLS else 1
    if index == len(S)-1:
        return a
    else:
        return (a+ c_or_v(S, index +1))
    
def check_vowells(S):
    ans = c_or_v(S)
    if ans>0:
        return (f'There are more consonants by {ans}')
    elif ans<0:
        return (f'There are more vowells by {ans}')
    else:
        return (f'There are an equal number of each {ans}')
    
    
strings_list = ['racecar', 'gohangasalamiimalasagnahog', 'notapalindrome', 'bob', 'a', 'hat', ]

for s in strings_list:
    print (check_vowells(s), s)
```

```python
#--------------C4-18-----------------
"""
Approaches:
1) Two lists that you join at the end
2) Add new even numbers to the front and odd ones to the back


"""

def evenoddbylists(S, index=0):
    if index == len(S)-1:
        if len(S) == 1:
            return S
        elif S[index]%2==0:
            return [S[index]], []
        else:
            return [], [S[index]]  
    else:
        evens, odds = evenoddbylists(S, index+1)
        if S[index] %2 == 0: evens.append(S[index])
        else: odds.append(S[index])
        if index == 0: return evens+odds
        else: return evens, odds
        
        
def evenoddbyappending(S, index=0):
    if index == len(S)-1:
        return [S[index]]
    
    else:
        if S[index]%2 == 1: #If odd, we want to add it to the end of the list
            return evenoddbyappending(S, index+1) + [S[index]]
        else:
            return [S[index]] + evenoddbyappending(S, index+1)
        
    
    
sequences = [[1,2,3,4,5,6,7,8], [4,3,65,23,5,46,765,3,45,23], [1], [2,2]]
print('Two lists approach')
for s in sequences:
    print(evenoddbylists(s))
    
    
print('\n\nAdding each element approach')
for s in sequences:
    print(evenoddbyappending(s))
    
    
"""
Final notes: the second approach is not desirable with lists because it is O(n) due to the fact that
appending an entire list of length m to another requires O(m) time.  This could be solved with Linked Lists when 
we get to that part of the book


Note also that both answers are correct, but the numbers are in a different order
"""
```

```python
#-------------C4-20--------------------
"""
Note: we can use the same approachs as before to do this in either O(n) time or O(n^2) time
Let's choose the O(n) approach

The major difference is that we need a third list for just k
"""

#There are issues adding a [a] + [], so this circumvents the problem
def add_lists(a,b):
    if a and b:
        return a+b
    else:
        return b or a #if one of them is [], return the other
    


def split_by_k(S, k, index=0):
    if len(S)==1:
        return S 
        
    elif index == len(S)-1:    #we use this line for too many of these exercises...
        if S[index] <k:
            return [S[index]], [], []
        elif S[index] >k:
            return [], [], [S[index]]
        else:
            return [], [S[index]], []  #Should return k in the middle list
        
    else:
        low, med, high = split_by_k(S, k, index+1)
        if S[index] <k:
            low = add_lists(low, [S[index]])
        elif S[index] >k:
            high = add_lists(high, [S[index]])
        else:
            med = add_lists(med, [S[index]])
            
        if index == 0:
            return add_lists(low, add_lists(med, high)) #This is not very elegant...
        else:
            return low, med, high
            

sequence = [11,2,33,4,5,6,7,8,9,10]

for i in range(0, 30, 2):
    print('Splitting by: ', i, 'produces', split_by_k(sequence, i))

"""
Note, it's easier to troubleshoot if you output the three lists seperately

This should run in O(n) time.
It relies on O(n) calls to the recursive function, each requiring O(1) time since we're adding at the end
The final concatenation of all the lists uses O(n) time, but this only happens once

Therefore O(n)*O(1) + O(n) is O(n)


"""
```

```python
#----------------C4-21----------------------
"""
Note: there is a brute force approach, but we will use the one with dictionaries instead


Additional Note: I learned an important thing about using mutable objects as defaults in python
Successive calls used the growing key_set, rather than initializing it as set() every time.

"""

#The following line is the wrong way to do the initialization...
#def find_pair(S, target, key_set = set(), index=0):

def find_pair(S, target, key_set = None, index=0):
    if key_set is None:
        key_set = set()
    
    
    key = target - S[index]
    if len(S) <= 1:
        return None
    
    elif index == len(S)-1:
        if key in key_set: return (key, S[index])
        else: return None
        
    else:
        if key in key_set: return (key, S[index])
        else:
            key_set.add(S[index])
            return find_pair(S, target, key_set, index+1)
        
    
sequence = [1,2,3,4,5,6,7,8]

for i in range(20):
    print(f'For the value {i}, the resulting pair is {find_pair(sequence, i)}')
```

```python
#--------------C4-22---------------------
"""
Note: this one took me a long time.  The secret is that we want to do the extra odd multiplications in reverse order
than you would get using the traditional:

    while factor>0:
        partial *= partial
        if factor%2 == 1:
            partial *= x
        factor = factor//2
    return partial
    
This will work in many cases, but not all.  Instead, you have to read in the sequence of odds backwards, which 
you can do using the bitwise notation (thanks to the hint in the book!)

Before that, you need to know how many relevant bits there are in your number, so the initial counter loop is for that


"""




def power_loop(x, n):
    factor = n
    partial = 1
    
    counter = 0
    while factor:
        factor = factor >>1
        counter += 1
    
    while counter+1:
        partial *= partial
        if n>>counter&1:
            partial *= x        
        counter -=1
    return partial
        
    
    
combos = [ (2,13), (2, 15), (3, 15), (10, 7), (2, 5)]
for b, p in combos:
    print('\n', power_loop(b, p), b**p)
        
        
    
```

```python
#------------P4-25-------------------
"""
The hint is that we should use a loop from 0 to 2**c -2
and that the number of ticks should be one more than the number of consecutive 1s.

That's a great hint!

"""

def english_ruler(n = 3, repeats = 10):
    limit = repeats*(2**n -2)
    counter = 0
    while counter<=limit:
        dashes = 1
        sub_counter = counter 
        while sub_counter & 1:
            dashes += 1
            sub_counter = sub_counter >> 1
        dashes = min (dashes, n)
        print('-'*dashes)
        counter += 1
        
english_ruler(4, 3)
            
    
```

```python
#----------------P4-27-------------------------
import os

def our_walk(path):
    folders = []
    files = []
    
    for obj in os.listdir(path):
        new_path = os.path.join(path, obj)
        if os.path.isdir(new_path):
            folders.append(obj)
            yield from our_walk (new_path)  #Note the use of yield from
        elif os.path.isfile(new_path):
            files.append(obj)
        else:
            print('Unknown object:', obj)
    
    yield(path, folders, files)



for i in os.walk('SampleFiles'):
    print(i)
    
print('\n')
for i in our_walk('SampleFiles'):
    print(i)
```
